```python
from openalgo import api

client = api(api_key="your-openalgo-apikey", host="http://127.0.0.1:5000")
```

### üß© 1. Single-Leg Option Example (Buy CE)

```python
response = client.margin(
    positions=[
        {
            "symbol": "NIFTY25NOV2525000CE",
            "exchange": "NFO",
            "action": "BUY",
            "product": "NRML",
            "pricetype": "MARKET",
            "quantity": "75",
        }
    ]
)
print(response)
```
